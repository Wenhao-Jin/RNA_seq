```python
import pandas as pd
import numpy as np
from matplotlib_venn import *
import matplotlib.pyplot as plt
```

```python
FUS_kd_gene=pd.read_table('FUS_kd/diff_out/gene_exp.diff')
FUS_human_gene=pd.read_table('FUS_human/diff_out/gene_exp.diff')
```

```python
##make the gene name in the table non-redundant: for those genes with more than one records (may result from the multi-loci), keep the record with highest fold-change
##for FUS_CTL table, also extract the nontg2homo part
FUS_kd_all_set=set(FUS_kd_gene['gene']) 
FUS_human_all_set=set(FUS_human_gene['gene'])
FUS_kd_gene_non_redun=pd.DataFrame(columns=FUS_kd_gene.columns)
FUS_non2homo_gene_non_redun=pd.DataFrame(columns=FUS_human_gene.columns)
for gene1 in FUS_kd_all_set:
    tmp_kd=FUS_kd_gene[FUS_kd_gene['gene']==gene1]
    if (len(tmp_kd)>1):#for those genes with more than one records (may result from the multi-loci), keep the record with highest fold-change
        tmp_kd=tmp_kd.sort('log2(fold_change)', ascending=False).iloc[0]
    FUS_kd_gene_non_redun.loc[len(FUS_kd_gene_non_redun)]=tmp_kd.iloc[0]
for gene2 in FUS_human_all_set:
    tmp_non2homo=FUS_human_gene[(FUS_human_gene['gene']==gene2) & (FUS_human_gene['sample_1']=='nontg') & (FUS_human_gene['sample_2']=='FUS-wt-homo')]
    if (len(tmp_non2homo)>1):#for those genes with more than one records (may result from the multi-loci), keep the record with highest fold-change
        tmp_non2homo=tmp_non2homo.sort('log2(fold_change)', ascending=False).iloc[0]
    if (len(tmp_non2homo)==0):
        continue
    FUS_non2homo_gene_non_redun.loc[len(FUS_non2homo_gene_non_redun)]=tmp_non2homo.iloc[0]
```

```python
FUS_non2homo_gene_non_redun
```

```python
##Differential expressed genes in FUS_kd dataset
FUS_kd_DEG=FUS_kd_gene_non_redun[(FUS_kd_gene_non_redun['q_value']<0.001) & ((FUS_kd_gene_non_redun['log2(fold_change)']>0.5850) | (FUS_kd_gene_non_redun['log2(fold_change)']<-0.5850))]
##Differential expressed genes in FUS_non2homo comparison
FUS_non2homo_DEG=FUS_non2homo_gene_non_redun[(FUS_non2homo_gene_non_redun['q_value']<0.001) & ((FUS_non2homo_gene_non_redun['log2(fold_change)']>0.5850) | (FUS_non2homo_gene_non_redun['log2(fold_change)']<-0.5850))]
kd_set=set(FUS_kd_DEG['gene'])         #using gene symbol to represent a record
#kd_set_loci=set(FUS_kd_DEG['locus'])   #using locus to represent a record
non2homo_set=set(FUS_non2homo_DEG['gene'])
#non2homo_set_loci=set(FUS_non2homo_DEG['locus'])
print('kd_DEG:'+str(len(FUS_kd_DEG)))
print('kd_DEG_up:'+str(len(FUS_kd_DEG[FUS_kd_DEG['log2(fold_change)']>0])))
print('kd_DEG_down:'+str(len(FUS_kd_DEG[FUS_kd_DEG['log2(fold_change)']<0])))
print('non2homo_DEG:'+str(len(FUS_non2homo_DEG)))
print('non2homo_DEG_up:'+str(len(FUS_non2homo_DEG[FUS_non2homo_DEG['log2(fold_change)']>0])))
print('non2homo_DEG_down:'+str(len(FUS_non2homo_DEG[FUS_non2homo_DEG['log2(fold_change)']<0])))
kd_non2homo_intersection=kd_set.intersection(non2homo_set)
#kd_non2homo_loci_intersection=kd_set_loci.intersection(non2homo_set_loci)
##get a table storing the common DEGs between FUS_kd and FUS_non2homo
FUS_kd_non2homo_ovlp_table=pd.DataFrame(columns=['gene','locus','log2(fc)_kd','log2(fc)_non2homo','q_val_kd','q_val_non2homo'])

##based on locus
#for locus in kd_non2homo_loci_intersection:
#    tmp=[np.array(FUS_kd_DEG['gene'][FUS_kd_DEG['locus']==locus])[0], locus, np.array(FUS_kd_DEG['log2(fold_change)'][FUS_kd_DEG['locus']==locus])[0], np.array(FUS_non2homo_DEG['log2(fold_change)'][FUS_non2homo_DEG['locus']==locus])[0], np.array(FUS_kd_DEG['q_value'][FUS_kd_DEG['locus']==locus])[0], np.array(FUS_non2homo_DEG['q_value'][FUS_non2homo_DEG['locus']==locus])[0]]
#    FUS_kd_non2homo_ovlp_table.loc[len(FUS_kd_non2homo_ovlp_table)]=tmp

##based on gene name
for gene in kd_non2homo_intersection:
    tmp_kd=FUS_kd_DEG[FUS_kd_DEG['gene']==gene]
    tmp_non2homo=FUS_non2homo_DEG[FUS_non2homo_DEG['gene']==gene]
    tmp=[gene, np.array(tmp_kd['locus'])[0], np.array(tmp_kd['log2(fold_change)'])[0], np.array(tmp_non2homo['log2(fold_change)'])[0], np.array(tmp_kd['q_value'])[0], np.array(tmp_non2homo['q_value'])[0]]
    FUS_kd_non2homo_ovlp_table.loc[len(FUS_kd_non2homo_ovlp_table)]=tmp

FUS_kd_non2homo_ovlp_upreg=FUS_kd_non2homo_ovlp_table[(FUS_kd_non2homo_ovlp_table['log2(fc)_kd']>0) & (FUS_kd_non2homo_ovlp_table['log2(fc)_non2homo']>0)].sort(['log2(fc)_kd','log2(fc)_non2homo'],ascending=False)
FUS_kd_non2homo_ovlp_downreg=FUS_kd_non2homo_ovlp_table[(FUS_kd_non2homo_ovlp_table['log2(fc)_kd']<0) & (FUS_kd_non2homo_ovlp_table['log2(fc)_non2homo']<0)].sort(['log2(fc)_kd','log2(fc)_non2homo'],ascending=True)
#FUS_kd_non2homo_ovlp_table
FUS_non2homo_DEG
```

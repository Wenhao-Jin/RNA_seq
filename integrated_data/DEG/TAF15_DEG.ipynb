```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import *
```

```python
##summary of the differential expressed genes
TAF15_htt2pbs_non_redun=pd.read_table('./TAF15/gene_htt2pbs_non_redun.xls')
TAF15_htt2kd_non_redun=pd.read_table('./TAF15/gene_htt2kd_non_redun.xls')
TAF15_pbs2kd_non_redun=pd.read_table('./TAF15/gene_pbs2kd_non_redun.xls')

##Differential expressed genes
TAF15_htt2pbs_DEG=TAF15_htt2pbs_non_redun[(TAF15_htt2pbs_non_redun['q_value']<0.01) & ((TAF15_htt2pbs_non_redun['log2(fold_change)']>0.5850) | (TAF15_htt2pbs_non_redun['log2(fold_change)']<-0.5850))]
TAF15_htt2kd_DEG=TAF15_htt2kd_non_redun[(TAF15_htt2kd_non_redun['q_value']<0.01) & ((TAF15_htt2kd_non_redun['log2(fold_change)']>0.5850) | (TAF15_htt2kd_non_redun['log2(fold_change)']<-0.5850))]
TAF15_pbs2kd_DEG=TAF15_pbs2kd_non_redun[(TAF15_pbs2kd_non_redun['q_value']<0.01) & ((TAF15_pbs2kd_non_redun['log2(fold_change)']>0.5850) | (TAF15_pbs2kd_non_redun['log2(fold_change)']<-0.5850))]

print('TAF15_htt2pbs_DEG:'+str(len(TAF15_htt2pbs_DEG)))
print('TAF15_htt2pbs_DEG_up:'+str(len(TAF15_htt2pbs_DEG[TAF15_htt2pbs_DEG['log2(fold_change)']>0])))
print('TAF15_htt2pbs_DEG_down:'+str(len(TAF15_htt2pbs_DEG[TAF15_htt2pbs_DEG['log2(fold_change)']<0])))
print('TAF15_htt2kd_DEG:'+str(len(TAF15_htt2kd_DEG)))
print('TAF15_htt2kd_DEG_up:'+str(len(TAF15_htt2kd_DEG[TAF15_htt2kd_DEG['log2(fold_change)']>0])))
print('TAF15_htt2kd_DEG_down:'+str(len(TAF15_htt2kd_DEG[TAF15_htt2kd_DEG['log2(fold_change)']<0])))
print('TAF15_pbs2kd_DEG:'+str(len(TAF15_pbs2kd_DEG)))
print('TAF15_pbs2kd_DEG_up:'+str(len(TAF15_pbs2kd_DEG[TAF15_pbs2kd_DEG['log2(fold_change)']>0])))
print('TAF15_pbs2kd_DEG_down:'+str(len(TAF15_pbs2kd_DEG[TAF15_pbs2kd_DEG['log2(fold_change)']<0])))

htt2pbs_set=set(TAF15_htt2pbs_DEG['gene'])
htt2kd_set=set(TAF15_htt2kd_DEG['gene'])
pbs2kd_set=set(TAF15_pbs2kd_DEG['gene'])
htt2pbs_htt2kd_intersection=htt2pbs_set.intersection(htt2kd_set)
htt2kd_pbs2kd_intersection=htt2kd_set.intersection(pbs2kd_set)
htt2pbs_pbs2kd_intersection=htt2pbs_set.intersection(pbs2kd_set)
#kd_non2homo_loci_intersection=kd_set_loci.intersection(non2homo_set_loci)
##get a table storing the common DEGs between FUS_kd and FUS_non2homo
htt2pbs_htt2kd_ovlp_table=pd.DataFrame(columns=['gene','locus','log2(fc)_htt2pbs','log2(fc)_htt2kd','q_val_htt2pbs','q_val_htt2kd'])
htt2kd_pbs2kd_ovlp_table=pd.DataFrame(columns=['gene','locus','log2(fc)_htt2kd','log2(fc)_pbs2kd','q_val_htt2kd','q_val_pbs2kd'])
htt2pbs_pbs2kd_ovlp_table=pd.DataFrame(columns=['gene','locus','log2(fc)_htt2pbs','log2(fc)_pbs2kd','q_val_htt2pbs','q_val_pbs2kd'])
##based on locus
#for locus in kd_non2homo_loci_intersection:
#    tmp=[np.array(FUS_kd_DEG['gene'][FUS_kd_DEG['locus']==locus])[0], locus, np.array(FUS_kd_DEG['log2(fold_change)'][FUS_kd_DEG['locus']==locus])[0], np.array(FUS_non2homo_DEG['log2(fold_change)'][FUS_non2homo_DEG['locus']==locus])[0], np.array(FUS_kd_DEG['q_value'][FUS_kd_DEG['locus']==locus])[0], np.array(FUS_non2homo_DEG['q_value'][FUS_non2homo_DEG['locus']==locus])[0]]
#    FUS_kd_non2homo_ovlp_table.loc[len(FUS_kd_non2homo_ovlp_table)]=tmp

##based on gene name, integrated kd and non2homo data into one table 
for gene in htt2pbs_htt2kd_intersection:
    tmp_htt2pbs=TAF15_htt2pbs_DEG[TAF15_htt2pbs_DEG['gene']==gene]
    tmp_htt2kd=TAF15_htt2kd_DEG[TAF15_htt2kd_DEG['gene']==gene]
    tmp=[gene, np.array(tmp_htt2pbs['locus'])[0], np.array(tmp_htt2pbs['log2(fold_change)'])[0], np.array(tmp_htt2kd['log2(fold_change)'])[0], np.array(tmp_htt2pbs['q_value'])[0], np.array(tmp_htt2kd['q_value'])[0]]
    htt2pbs_htt2kd_ovlp_table.loc[len(htt2pbs_htt2kd_ovlp_table)]=tmp
for gene in htt2kd_pbs2kd_intersection:
    tmp_htt2kd=TAF15_htt2kd_DEG[TAF15_htt2kd_DEG['gene']==gene]
    tmp_pbs2kd=TAF15_pbs2kd_DEG[TAF15_pbs2kd_DEG['gene']==gene]
    tmp=[gene, np.array(tmp_htt2kd['locus'])[0], np.array(tmp_htt2kd['log2(fold_change)'])[0], np.array(tmp_pbs2kd['log2(fold_change)'])[0], np.array(tmp_htt2kd['q_value'])[0], np.array(tmp_pbs2kd['q_value'])[0]]
    htt2kd_pbs2kd_ovlp_table.loc[len(htt2kd_pbs2kd_ovlp_table)]=tmp
for gene in htt2pbs_pbs2kd_intersection:
    tmp_htt2pbs=TAF15_htt2pbs_DEG[TAF15_htt2pbs_DEG['gene']==gene]
    tmp_pbs2kd=TAF15_pbs2kd_DEG[TAF15_pbs2kd_DEG['gene']==gene]
    tmp=[gene, np.array(tmp_htt2pbs['locus'])[0], np.array(tmp_htt2pbs['log2(fold_change)'])[0], np.array(tmp_pbs2kd['log2(fold_change)'])[0], np.array(tmp_htt2pbs['q_value'])[0], np.array(tmp_pbs2kd['q_value'])[0]]
    htt2pbs_pbs2kd_ovlp_table.loc[len(htt2pbs_pbs2kd_ovlp_table)]=tmp
htt2pbs_htt2kd_ovlp_upreg=htt2pbs_htt2kd_ovlp_table[(htt2pbs_htt2kd_ovlp_table['log2(fc)_htt2pbs']>0) & (htt2pbs_htt2kd_ovlp_table['log2(fc)_htt2kd']>0)].sort(['log2(fc)_htt2pbs','log2(fc)_htt2kd'],ascending=False)
htt2pbs_htt2kd_ovlp_downreg=htt2pbs_htt2kd_ovlp_table[(htt2pbs_htt2kd_ovlp_table['log2(fc)_htt2pbs']<0) & (htt2pbs_htt2kd_ovlp_table['log2(fc)_htt2kd']<0)].sort(['log2(fc)_htt2pbs','log2(fc)_htt2kd'],ascending=True)
htt2kd_pbs2kd_ovlp_upreg=htt2kd_pbs2kd_ovlp_table[(htt2kd_pbs2kd_ovlp_table['log2(fc)_htt2kd']>0) & (htt2kd_pbs2kd_ovlp_table['log2(fc)_pbs2kd']>0)].sort(['log2(fc)_htt2kd','log2(fc)_pbs2kd'],ascending=False)
htt2kd_pbs2kd_ovlp_downreg=htt2kd_pbs2kd_ovlp_table[(htt2kd_pbs2kd_ovlp_table['log2(fc)_htt2kd']<0) & (htt2kd_pbs2kd_ovlp_table['log2(fc)_pbs2kd']<0)].sort(['log2(fc)_htt2kd','log2(fc)_pbs2kd'],ascending=True)
htt2pbs_pbs2kd_ovlp_upreg=htt2pbs_pbs2kd_ovlp_table[(htt2pbs_pbs2kd_ovlp_table['log2(fc)_htt2pbs']>0) & (htt2pbs_pbs2kd_ovlp_table['log2(fc)_pbs2kd']>0)].sort(['log2(fc)_htt2pbs','log2(fc)_pbs2kd'],ascending=False)
htt2pbs_pbs2kd_ovlp_downreg=htt2pbs_pbs2kd_ovlp_table[(htt2pbs_pbs2kd_ovlp_table['log2(fc)_htt2pbs']<0) & (htt2pbs_pbs2kd_ovlp_table['log2(fc)_pbs2kd']<0)].sort(['log2(fc)_htt2pbs','log2(fc)_pbs2kd'],ascending=True)
print('htt2pbs_htt2kd_ovlp_upreg:'+str(len(htt2pbs_htt2kd_ovlp_upreg)))
print('htt2pbs_htt2kd_ovlp_downreg:'+str(len(htt2pbs_htt2kd_ovlp_downreg)))
print('htt2pbs_pbs2kd_ovlp_upreg:'+str(len(htt2pbs_pbs2kd_ovlp_upreg)))
print('htt2pbs_pbs2kd_ovlp_downreg:'+str(len(htt2pbs_pbs2kd_ovlp_downreg)))
print('htt2kd_pbs2kd_ovlp_upreg:'+str(len(htt2kd_pbs2kd_ovlp_upreg)))
print('htt2kd_pbs2kd_ovlp_downreg:'+str(len(htt2kd_pbs2kd_ovlp_downreg)))
```
